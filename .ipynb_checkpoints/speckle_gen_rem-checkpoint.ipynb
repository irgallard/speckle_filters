```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage

#Fontstyle for the plots
font = {'family': 'serif',
    'color':  'black',
    'weight': 'normal',
    'size': 16,
    }
```

```python
#Load and display the image
path = 's1a-wv1-slc-vv-20160323t140548-20160323t140551-010497-00f963-111.png'
#path = 'Guia-de-Genshin-Impact-Yae-Miko-materiales-mejor-compilacion-para.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


def display_image(img, cat, font):
    """
    Function to display a SAR image and return it
    Arguments:
        path: path to the image file
        font: for pretty plots
    Rerturns:
        img: image file
    """
    plt.xlabel("Width", fontdict = font)
    plt.ylabel("Height", fontdict = font)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    #plt.title("Original image", fontdict = font)
    plt.imshow(img, cmap = plt.cm.gray)
    if cat == "original":
        plt.title("{} image".format(cat.capitalize()), fontdict = font)
    if cat == "speckled":
        plt.title("{} image".format(cat.capitalize()), fontdict = font)
    if cat == "filtered":
        plt.title("{} image".format(cat.capitalize()), fontdict = font)
    return(img)

cat = "original"
normal_image = display_image(img, cat, font)
print("Image shape: ", normal_image.shape)
```

```python
#Speckle noise mode, see paper
#Skimage has an implementation too

def add_speckle(image):
    gauss = np.random.normal(0,1,image.size)
    gauss = gauss.reshape(image.shape[0], image.shape[1]).astype('uint8')
    return np.multiply(image,gauss) + image

cat = "speckled"
noisy_image = add_speckle(normal_image)
noisy_display = display_image(noisy_image, cat,font)
```

```python
#Scipy ndimage has an implementation of this, results are the same
def median_filter(img, ksize):
    """
    Arguments:
        img: the  noisy image
        ksize: kernel size, integer
    Returns:
        filtered_image: image after median filter
    """
    #Pad the image edges
    pad_size = ksize//2
    img_padded = np.pad(img, pad_size, mode='constant')
    #Empty array for the filtered image
    filtered_img = np.empty_like(img)
    # Iterate over the image
    width = img.shape[0]
    height = img.shape[1]
    for i in range(width):
        for j in range(height):
            #Store the local window
            window = img_padded[i:i+ksize, j:j+ksize]
            # Calculate the median of the pixels in the window
            median = np.median(window)
            # Assign the median value to the filtered image array
            filtered_img[i, j] = median
    return filtered_img


#Apply the filter to the noisy image
fixed_image = median_filter(noisy_image, 3)

cat = "filtered"
fixed_image = display_image(fixed_image, cat, font)
```

```python
def lee_filter(img, window_size=3):
    """
    Applies the Lee filter to the input image to reduce speckle noise.
    
    Parameters:
    - img: The input image (grayscale or multichannel) as a 2D NumPy array.
    - window_size: The size of the local window used in the filter (default is 3).
    
    Returns:
    - The filtered image as a 2D NumPy array.
    """
    # Initialize the filtered image with the same shape as the input image
    filtered_img = np.zeros_like(img)
    
    # Iterate over the image pixels
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # Compute the indices of the local window
            i_min = max(i - window_size // 2, 0)
            i_max = min(i + window_size // 2 + 1, img.shape[0])
            j_min = max(j - window_size // 2, 0)
            j_max = min(j + window_size // 2 + 1, img.shape[1])
            
            # Extract the local window
            window = img[i_min:i_max, j_min:j_max]
            
            # Compute the mean and standard deviation of the local window
            mean = np.mean(window)
            std = np.std(window)
            
            # Compute the filtered pixel value using the Lee filter equation
            filtered_img[i, j] = mean + (img[i, j] - mean) * (1 - std / (std + 1))
    
    return filtered_img
```
