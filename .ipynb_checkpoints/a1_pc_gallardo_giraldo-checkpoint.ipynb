```python
#Imports needed to load, do math and plot
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Fontstyle for the plots
font = {'family': 'serif',
    'color':  'black',
    'weight': 'normal',
    'size': 16,
    }
```

```python
#Load the image, convert it to gray and resize
path = 'city.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (400,400))


#Function to display any single image
def display_image(img, cat, font):
    """
    Function to display an image and return it
    Arguments:
        path: path to the image file
        cat: string used for the title
        font: for pretty plots
    Rerturns:
        img: image file
    """
    #Plot
    plt.xlabel("Width", fontdict = font)
    plt.ylabel("Height", fontdict = font)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.imshow(img, cmap = plt.cm.gray)
    #Choose the title depending on cat
    if cat == "original":
        plt.title("{} image".format(cat.capitalize()), fontdict = font)
    if cat == "speckle":
        plt.title("Added {} noise".format(cat.capitalize()), fontdict = font)
    if cat == "median":
        plt.title("After {} filter".format(cat.capitalize()), fontdict = font)
    if cat == "lee":
        plt.title("After {} filter".format(cat.capitalize()), fontdict = font)
    return(img)

cat = "original"
normal_image = display_image(img, cat, font)
```

```python
#Speckle noise implementation from the papers. Skimage has one too.
#Usual in SAR images. Similar to quantum mottle noise in radiographies

def add_speckle(image, mean, var):
    """
    Add speckle noise to an image.
    Arguments:
        image: the loaded image
        mean: mean of ther gaussian distributions for random values
        var: its variance
    Returns:
        noisy: the image with speckle noise
    """
    #Gaussian noise
    gauss = np.random.normal(mean,np.sqrt(var),image.size)
    gauss = gauss.reshape(image.shape[0], image.shape[1]).astype('uint8')
    noisy = np.multiply(image,gauss) + image
    return noisy

#Play with the mean and variance parameters for different effects
noisy_image = add_speckle(normal_image, mean =  0.25, var = 0.3)
```

```python
#Display the noisy image
cat = "speckle"
noisy_display = display_image(noisy_image, cat,font)
```

```python
#Implementation of median filter
#Works pretty well for different params Speckle noise
def median_filter(img, ksize):
    """
    Arguments:
        img: the  noisy image
        ksize: kernel size, integer
    Returns:
        filtered_image: image after median filter
    """
    #Pad the image edges
    pad_size = ksize//2
    img_padded = np.pad(img, pad_size, mode='constant')
    #Empty array for the filtered image
    filtered_img = np.empty_like(img)
    # Iterate over the image
    width = img.shape[0]
    height = img.shape[1]
    for i in range(width):
        for j in range(height):
            #Store the local window
            window = img_padded[i:i+ksize, j:j+ksize]
            # Calculate the median of the pixels in the window
            median = np.median(window)
            # Assign the median value to the filtered image array
            filtered_img[i, j] = median
    return filtered_img


#Apply the filter to the noisy image
#Kernel size 3 for SAR images, play with it
fixed_image = median_filter(noisy_image, 3)


```

```python
#Display the median filtered image
cat = "median"
median_image = display_image(fixed_image, cat, font)
```

```python
#Implementation of a hybrid median + Lee filter
#Works well for certain parameters Speckle noise and SAR images
#Usually reduces less noise but keeps more information
def lee_filter(img, ksize):
    """
    Arguments:
        img: the  noisy image
        ksize: kernel size, integer
    Returns:
        filtered_image: image after Lee filter
    """
    #Add padding 
    pad_size = ksize//2
    img_padded = np.pad(img, pad_size, mode='constant')
    #Empty array for the filtered image
    filtered_img = np.empty_like(img)
    #Variance of the image
    std0 = np.std(img)
    # Iterate over the image
    width = img.shape[0]
    height = img.shape[1]
    #Small number to avoid 0 division
    eta = 0.000001
    for i in range(width):
        for j in range(height):
            #Store the local window
            window = img_padded[i:i+ksize, j:j+ksize]
            # Calculate the equation params
            mean = np.mean(window)
            std = np.std(window)
            W = (std**2)/(std**2+std0**2+eta)
            newpix = mean + W*(img[i,j]-mean)
            median = np.median(window)
            val = np.abs(median-newpix)
            if val >= 20:
                newpix = median
            filtered_img[i, j] = newpix
    return filtered_img

#Change kernel size depending on the image
fixed_image = lee_filter(noisy_image, ksize = 3)
```

```python
#Display the Lee filtered image
cat = "lee"
lee_image = display_image(fixed_image, cat, font)
```

```python
#Plot comparison for original, noisy, median and lee
plt.figure(figsize=(20,10))

plt.subplot(1, 4, 1)
plt.imshow(normal_image, cmap = 'gray')
plt.title('Original', fontdict =font)
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(noisy_image, cmap = 'gray')
plt.axis('off')
plt.title('Ruido de moteado',  fontdict =font)

plt.subplot(1, 4, 3)
plt.imshow(median_image, cmap = 'gray')
plt.axis('off')
plt.title('Filtro de mediana', fontdict =font)

plt.subplot(1, 4, 4)
plt.imshow(lee_image, cmap = 'gray')
plt.axis('off')
plt.title('Filtro hbrido Lee', fontdict =font);



```

```python
#Histogram eq. for radiographies or other images

def hist_eq(img):
    """
    Arguments:
        img: img to equalize
    Returns:
        eq_img: equalized image
    """
    #All arguments are lists or np arrays
    #Args: img, channels, mask, histSize, ranges
    N = 512
    hist = cv2.calcHist([img], [0], None, [N], [0, N])
    #Normalize histogram
    normalized = hist/img.size
    #New np array for the equalized image
    width, height = img.shape
    eq_img = np.ones((width, height), np.uint8)
    #Cumulative and loop
    cumulative = normalized.cumsum()*255
    for i in range(width):
        for j in range(height):
            eq_img[i, j] = cumulative[img[i, j]]
    return eq_img
        
eq_img = hist_eq(normal_image)
```

```python
#Plot comparison between original, noisy, median filter and median + hist
plt.figure(figsize=(20, 10))

plt.subplot(1, 4, 1)
plt.imshow(normal_image, cmap = 'gray')
plt.title('Original', fontdict =font)
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(noisy_image, cmap = 'gray')
plt.axis('off')
plt.title('Ruido de moteado',  fontdict =font)

plt.subplot(1, 4, 3)
plt.imshow(median_image, cmap = 'gray')
plt.axis('off')
plt.title('FIltro de mediana', fontdict =font)

plt.subplot(1, 4, 4)
plt.imshow(eq_img, cmap = 'gray')
plt.axis('off')
plt.title('Filtro mediana + Histograma', fontdict =font);
```
