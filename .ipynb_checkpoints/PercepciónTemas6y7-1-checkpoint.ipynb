# Ley de potencia - Transformación de Gamma

```python
# ///////////////////////
# Correccion gama
# Autor: Bryan Medina
# ///////////////////////

import cv2
import numpy as np
import matplotlib.pyplot as plt

# ubicación de la imagen
f = cv2.imread('lena_square_half.png', 0)

#asignamos gamma y obtenemos las constantes c encargadas de mantener a raya los valores del despliegue (0 - 255)
# en este caso btenemos la función que graficaremos.

gamma = 0.4
x = np.array(range(256))
#np.power elementos de la primera matriz elevados a potenciales de la segunda.
c = 255/np.power(255, gamma)
y =  c * np.power(x, gamma)

gamma1 = 1.0
c1 = 255/np.power(255, gamma1)
y1 =  c1 * np.power(x, gamma1)

gamma2 = 2.5

c2 = 255/np.power(255, gamma2)
y2 =  c2 * np.power(x, gamma2)

#graficamos las funciones

plt.figure()
plt.title('correción gamma')
plt.plot(x, y, label = '$\gamma = 0.4$')
plt.plot(x, y1, label = '$\gamma = 1.0$')
plt.plot(x, y2, label = '$\gamma = 2.5$')
plt.legend(loc='upper left')
plt.xlabel('intensidades de entrada')
plt.ylabel('intensidades de salida')
plt.show()

#calculamos la transformacion gamma
outputImg = np.array(c * np.power(f, gamma), dtype = 'uint8')
outputImg1 = np.array(c2 * np.power(f, gamma2), dtype = 'uint8')


cv2.imshow('gamma = 0.5', outputImg)
cv2.imshow('gamma = 2.0', outputImg1)
cv2.imshow('entrada', f)
#cv2.imwrite('punto5.jpg', outputImg)
#cv2.imwrite('2puntocero.jpg', outputImg1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Equalización de histograma

```python
#calculo del histograma de la imagen de entrada f
#Los histogramas nos dan la distribución de intensidad de los píxeles y prevalecen en casi todos los aspectos de la 
#visión artificial.

#Usamos histogramas en escala de grises para el umbral, para balancear blancos y para el seguimiento de objetos en imágenes.

#cv2.calcHist(images, channels, mask, histSize, range)
#channels: Para calcular un histograma de una imagen en escala de grises, la lista sería [0]. 
 # Para calcular un histograma para los tres canales rojo, verde y azul, la lista de canales sería [0, 1, 2].
#mask: En este link podemos encontrar explicación detallada de las máscaras: 
# https://pyimagesearch.com/2021/01/19/image-masking-with-opencv/
#histSize: este es el número de contenedores que queremos usar al calcular un histograma.


histF = cv2.calcHist([f], [0], None, [256], [0, 256])
x = np.array(range(256))

#número de elementos de la imagen, para normalizar
numel = f.size

#histograma normalizado (0 - 1) (divido entre todos los elementos)
histNorm = histF/numel

#suma acumulada del histograma normalizado
##Genera la función de distribución acumulada (cdf por sus siglas en inglés cumulative distribution function)

# cdf = np.sum(histNorm)
cdf = histNorm.cumsum()
cdf = cdf * 255


width, height = f.shape
g = np.zeros((width, height), np.uint8)

#Para entenderlo mejor podemos ir al siguiente link: https://www.datasmarts.net/ecualizacion-de-histograma-con-opencv/

for i in range(width):
    for j in range(height):

        g[i, j] = cdf[f[i, j]]

        

cv2.imwrite('lena_histograma.jpg', g)

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(f, cmap = 'gray')
plt.title('Imagen de entrada')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(g, cmap = 'gray')
plt.axis('off')
plt.title('Ecualización de histograma')
plt.show()


histG = cv2.calcHist([g], [0], None, [256], [0, 256])


plt.figure()
plt.subplot(1, 3, 1)
plt.title('Histograma')
plt.plot(x, histF)

plt.subplot(1, 3, 2)
plt.title('Dens. Acumulada')
plt.plot(x, cdf)

plt.subplot(1, 3, 3)
plt.title('Hist. Ecualización')
plt.stem(x, histG)
plt.show()
```

# Filtro paso alto

```python
import cv2
import numpy as np
from scipy import ndimage

kernel_3x3 = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1, 1, 2, 2, -1],
                       [-1, 2, 4, 2, -1],
                       [-1, 1, 2, 1, -1],
                       [-1, -1, -1, -1, -1]])

img = cv2.imread("lena_square_half.png", cv2.IMREAD_GRAYSCALE)

k3 = ndimage.convolve(img, kernel_3x3)
k5 = ndimage.convolve(img, kernel_5x5)

blurred = cv2.GaussianBlur(img, (11, 11), 0)
g_hpf = img - blurred

cv2.imshow("3x3", k3)
cv2.imshow("5x5", k5)
cv2.imshow("g_hpf", g_hpf)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Filtro paso bajo

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('lena_square_half.png')
 
#Crea el kernel
kernel = np.ones((5,5),np.float32)/25
kernel3 = np.ones((3,3),np.float32)/9
 
#Filtra la imagen utilizando el kernel anterior
img_pasoBajo5 = cv2.filter2D(img,-1,kernel)
img_pasoBajo3 = cv2.filter2D(img,-1,kernel3)
 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_pasoBajo5),plt.title('Filtro paso bajo 5')
plt.xticks([]), plt.yticks([])

plt.show()

```

# Filtro canny

```python
# Filtro Gaussiano --> Gradiente --> Algoritmo de histéresis.
canny = cv2.Canny(img, 100,200)
cv2.imshow('Img', canny)
cv2.imshow('Canny - valMin =100 - valMax = 200', canny)

cv2.waitKey(0)
cv2.destroyAllWindows()


#ALGORITMO CANNY
# 1. utiliza un filtro Gaussiano para suavizar la imagen. 
# 2. Encuentra el gradiente. El gradiente define dos valores: la dirección en donde el cambio de intensidad es máximo 
# y la magnitud de esa dirección.
# 3. Después de obtener la magnitud y la dirección del degradado, se realiza un escaneo completo de la imagen para eliminar 
# los píxeles no deseados que pueden no constituir el borde. Esto elimina los píxeles que no se consideran parte de un borde. 
# Por lo tanto, solo quedarán líneas finas (bordes candidatos).
# 4. Esta etapa decide cuáles son los bordes que son realmente bordes y cuáles no. 
# Para esto, necesitamos dos valores de umbral, minVal y maxVal.

#ALGORITMO HISTÉRESIS utilizado para detectar esos valores mínimos y máximos de intensidades
#1. Si el píxel tiene intensidad 0 (negro), no pertenece a ningún grupo pues no es un componente a etiquetar.
#2. Si tiene un valor distinto de 0: Si el píxel superior, izquierdo, superior izquierdo o superior derecho tienen intensidad cero, este píxel es marcado
#   como perteneciente a nuevo grupo.
#3. Si tiene un valor distinto de 0: Si el píxel superior, izquierdo, superior izquierdo o superior derecho tienen intensidad mayor a cero, entonces
#   aquellos con intensidad mayor a cero ya fueron recorridos y catalogados en un grupo.
#4. Si todos estos ”vecinos” con intensidad mayor a cero son del mismo grupo N, este píxel en catalogado como grupo N.
#5. Si existen conflictos, se lo cataloga como grupo de alguno de los vecinos, y los demás grupos son marcados como ”en conflicto”.
```
